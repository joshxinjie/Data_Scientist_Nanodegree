`1.` Above, you can see the actual and predicted columns for each of the datasets.  Using the **jud_data**, find the proportion of errors for the dataset, and furthermore, the percentage of errors of each type.  Use the results to answer the questions in quiz 1 below.  

**Hint for quiz:** an error is any time the prediction doesn't match an actual value.  Additionally, there are Type I and Type II errors to think about.  We also know we can minimize one type of error by maximizing the other type of error.  If we predict all individuals as innocent, how many of the guilty are incorrectly labeled?  Similarly, if we predict all individuals as guilty, how many of the innocent are incorrectly labeled?

```python
# Number of errors
num_err = jud_data[jud_data['actual'] != jud_data['predicted']].shape[0]
err_per = num_err / jud_data.shape[0]
err_per
```

```python
# Type 1 error
false_positives = jud_data[(jud_data['actual'] == 'innocent') &\
                           (jud_data['predicted'] == 'guilty')].shape[0]
type_1_err = false_positives/jud_data.shape[0]
type_1_err
```

```python
# Type 2 error
false_negatives = jud_data[(jud_data['actual'] == 'guilty') &\
                           (jud_data['predicted'] == 'innocent')].shape[0]
type_2_err = false_negatives/jud_data.shape[0]
type_2_err
```

If we predict all individuals to be guilty, then there will be no one who is predicted innocent. Then type 2 error will be 0

```python
# If every one is predicted guilty, then type 1 error will be number of
# innocents over total number of defendants

jud_data[jud_data['actual'] == 'innocent'].shape[0]/jud_data.shape[0]
```

`2.` Using the **par_data**, find the proportion of errors for the dataset, and furthermore, the percentage of errors of each type.  Use the results to answer the questions in quiz 2 below.

These should be very similar operations to those you performed in the previous question.

```python
# Number of errors
num_err = par_data[par_data['actual'] != par_data['predicted']].shape[0]
err_per = num_err / par_data.shape[0]
err_per
```

```python
# Type 1 error
false_positives = par_data[(par_data['actual'] == 'fails') &\
                           (par_data['predicted'] == 'opens')].shape[0]
type_1_err = false_positives/par_data.shape[0]
type_1_err
```

```python
# Type 2 error
false_positives = par_data[(par_data['actual'] == 'opens') &\
                           (par_data['predicted'] == 'fails')].shape[0]
type_2_err = false_positives/par_data.shape[0]
type_2_err
```

if every parachute is predicted to fail, then no parachute will be predicted to open, hence type 1 error is 0

```python
# if every parachute is predicted to fail, then type 2 error will be
# number of parachutes that open over total number of parachutes

par_data[par_data['actual'] == 'opens'].shape[0]/par_data.shape[0]
```
