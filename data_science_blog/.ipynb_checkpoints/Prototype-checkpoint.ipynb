## Build training/test data for regression model

```python
bos_listings_reg = bos_listings[['id', 'space', 'neighborhood_overview', 'notes', 'transit',\
                                 'access', 'interaction', 'house_rules', 'thumbnail_url', 'medium_url',\
                                 'host_about', 'host_response_time', 'host_response_rate',\
                                 'host_acceptance_rate', 'host_is_superhost', 'host_identity_verified',\
                                 'neighbourhood_cleansed', 'is_location_exact', 'property_type',\
                                 'room_type', 'accommodates', 'bathrooms', 'bedrooms', 'beds',\
                                 'bed_type', 'amenities', 'price', 'cleaning_fee', 'extra_people',\
                                 'number_of_reviews',\
                                 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',\
                                 'review_scores_checkin', 'review_scores_communication',\
                                 'review_scores_location', 'review_scores_value', 'instant_bookable',\
                                 'cancellation_policy', 'require_guest_profile_picture',\
                                 'require_guest_phone_verification', 'reviews_per_month']]

# Rename id to listing_id so that we can concate this with other dataframes
bos_listings_reg.rename(index=str, columns={"id": "listing_id"},inplace=True)

bos_listings_reg.head()
```

```python
missing_per_row = bos_listings_reg.isnull().mean()
missing_per_row
```

```python
# Engineer new boolean variables to keep track if variable exits
nan_bool_var = ['space', 'neighborhood_overview', 'notes', 'transit', 'access', 'interaction',\
                'house_rules', 'thumbnail_url', 'medium_url', 'host_about','amenities']

for var in nan_bool_var:
    bool_var_name = 'has_' + var
    bos_listings_reg[bool_var_name] = np.where(bos_listings_reg[var].isnull(), 0, 1)
    
# Drop nan_bool_var
bos_listings_reg.drop(columns=nan_bool_var, inplace=True)
    
bos_listings_reg.head()
```

```python
#check
bos_listings_reg.has_space.value_counts(dropna=False)
```

```python
# Label Encode true/false variables
var_to_lab_enc = ['host_is_superhost', 'host_identity_verified', 'is_location_exact',\
                  'instant_bookable', 'require_guest_profile_picture', 'require_guest_phone_verification']

for var in var_to_lab_enc:
    bos_listings_reg[var] = np.where(bos_listings_reg[var] == "t", 1, 0)

bos_listings_reg.head()
```

```python
#check
bos_listings_reg.host_identity_verified.value_counts(dropna=False)
```

```python
bos_listings.host_identity_verified.value_counts(dropna=False)
```

```python
# Engineer Categorical variables
# also engineer an nan column for each of them
# use pandas get_dummies
cat_var_to_enc = ['host_response_time', 'neighbourhood_cleansed', 'property_type', 'room_type',\
                  'accommodates', 'bathrooms', 'bedrooms', 'beds', 'bed_type', 'cancellation_policy']

for var in cat_var_to_enc:
    bos_listings_reg = pd.concat([bos_listings_reg.drop(var, axis=1),\
                                     pd.get_dummies(bos_listings_reg[var],\
                                                    prefix=var, prefix_sep='_', drop_first=False, dummy_na=True)], axis=1)

bos_listings_reg.head()
```

```python
from bokeh.io import push_notebook, show, output_notebook, output_file
from bokeh.models import GMapPlot, GMapOptions, ColumnDataSource, Circle, Range1d, BasicTicker, ColorBar, PanTool, WheelZoomTool, BoxSelectTool
from bokeh.models.mappers import ColorMapper, LinearColorMapper
from bokeh.palettes import inferno
```

```python
#from bokeh.palettes import magma
#magma(25)
```

```python
# Can change size to number of available bookings

output_file("bos_nigh_price.html")

map_options = GMapOptions(lat=42.3601, lng=-71.0589, map_type="roadmap", zoom=11)

plot = GMapPlot(api_key=MY_API_KEY, x_range=Range1d(),\
                y_range=Range1d(), map_options=map_options,\
                width=1200, height=750)

plot.title.text = "Boston"

source = ColumnDataSource(
    data=dict(
        lat=bos_neigh_price_df['latitude'],\
        lon=bos_neigh_price_df['longitude'],\
        size=bos_neigh_price_df['price'] / 5,\
        color=bos_neigh_price_df['price']
    )
)

color_mapper = LinearColorMapper(palette=inferno(25), low=bos_neigh_price_df['price'].min(), high=bos_neigh_price_df['price'].max())

circle = Circle(x="lon", y="lat", size="size", fill_color={'field': 'color', 'transform': color_mapper}, fill_alpha=0.7, line_color=None)

plot.add_glyph(source, circle)

color_bar = ColorBar(color_mapper=color_mapper, ticker=BasicTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))
plot.add_layout(color_bar, 'right')

plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool())

show(plot)
```
